## Calculate the amount of water flowing over a time period



The [flow rate](https://sutherland.che.utah.edu/wiki/index.php/Numerical_Integration#Example_2) in a river on each day is tabulated below.

| Day|Flowrate (CFS)|
|---|---|
| 1|10000|
| 2|12000|
| 3|15000|
| 4|13500|
| 5|14400|

Make a plot of the data.



To calculate the total amount of water that flowed through the river over the 5 days you need to integrate the data above. Use `np.trapz` to do that.
